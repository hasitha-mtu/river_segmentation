```python
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv(image_stats_file)

# Identify all Dice columns
dice_cols = [col for col in df.columns if col.startswith('dice_')]

# Calculate Consensus Dice (row-wise mean)
df['consensus_dice'] = df[dice_cols].mean(axis=1)
```

```python
df
```

```python
# Shapiro-Wilk for Normality
_, p_dice = stats.shapiro(df['consensus_dice'])
_, p_ratio = stats.shapiro(df['gt_ratio'])

# Spearman Correlation (Consensus)
rho_consensus, p_val = stats.spearmanr(df['gt_ratio'], df['consensus_dice'])

# Per-Model Spearman (Range and Average)
# Iterate through each model's specific dice column
model_rhos = [stats.spearmanr(df['gt_ratio'], df[col])[0] for col in dice_cols]

print(f"Consensus Rho: {rho_consensus:.3f}")
print(f"Mean Per-Model Rho: {np.mean(model_rhos):.3f} (Range: {min(model_rhos):.3f}-{max(model_rhos):.3f})")
```

```python
y = df['consensus_dice'].values.reshape(-1, 1)
x_lin = df['gt_ratio'].values.reshape(-1, 1)
x_log = np.log(df['gt_ratio']).values.reshape(-1, 1)

# Fit Linear
reg_lin = LinearRegression().fit(x_lin, y)
r2_lin = r2_score(y, reg_lin.predict(x_lin))

# Fit Log-Linear
reg_log = LinearRegression().fit(x_log, y)
r2_log = r2_score(y, reg_log.predict(x_log))

print(f"Linear R2: {r2_lin:.3f}, Log-Linear R2: {r2_log:.3f}")
```

```python
# Calculate residuals from the linear model
df['residuals'] = y - reg_lin.predict(x_lin)

# Sort by gt_ratio and split into thirds
df_sorted = df.sort_values('gt_ratio')
low, mid, high = np.array_split(df_sorted, 3)

print(f"Residual Std (Low Coverage): {low['residuals'].std():.3f}")
print(f"Residual Std (Mid Coverage): {mid['residuals'].std():.3f}")
```

```python
# Sort by filename to restore flight sequence
df = df.sort_values('image')

# Calculate correlation between row i and row i-1
lag_1 = df['consensus_dice'].autocorr(lag=1)

print(f"Lag-1 Autocorrelation: {lag_1:.2f}")
```

```python
from scipy import stats
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(image_stats_file)

# 2. Identify Dice columns and calculate Consensus Dice
dice_cols = [col for col in df.columns if col.startswith('dice_')]
df['consensus_dice'] = df[dice_cols].mean(axis=1)

# 3. Define the Partial Spearman Function
# This calculates the correlation between X and Y while removing the influence of Z
def partial_spearman(df, x_col, y_col, z_col):
    # Step 1: Rank the data (Spearman is Pearson on ranks)
    df_ranked = df[[x_col, y_col, z_col]].rank()
    
    # Step 2: Get residuals of X vs Z
    reg_x = LinearRegression().fit(df_ranked[[z_col]], df_ranked[x_col])
    res_x = df_ranked[x_col] - reg_x.predict(df_ranked[[z_col]])
    
    # Step 3: Get residuals of Y vs Z
    reg_y = LinearRegression().fit(df_ranked[[z_col]], df_ranked[y_col])
    res_y = df_ranked[y_col] - reg_y.predict(df_ranked[[z_col]])
    
    # Step 4: Pearson correlation of residuals = Partial Spearman
    rho, p = stats.pearsonr(res_x, res_y)
    return rho, p

# --- VALIDATION CALCULATIONS ---

# 1. Laplacian Variance Analysis
raw_rho, _ = stats.spearmanr(df['laplacian_var'], df['consensus_dice'])
partial_rho, partial_p = partial_spearman(df, 'laplacian_var', 'consensus_dice', 'gt_ratio')

print(f"Laplacian Raw Rho: {raw_rho:.3f}")
print(f"Laplacian Partial Rho (adj. for gt_ratio): {partial_rho:.3f} (p={partial_p:.3f})")

# 2. High Saturation Analysis
sat_rho, sat_p = stats.spearmanr(df['high_sat_frac'], df['consensus_dice'])
sat_collinearity, _ = stats.spearmanr(df['high_sat_frac'], df['gt_ratio'])

print(f"High Sat Rho: {sat_rho:.3f} (p={sat_p:.3f})")
print(f"High Sat Collinearity with gt_ratio: {sat_collinearity:.3f}")

# 3. Other Statistics (Loop for efficiency)
others = {
    'Dark Pixels': 'low_val_frac',
    'ExG Mean': 'exg_mean',
    'HSV Value Std': 'hsv_v_std'
}

for label, col in others.items():
    rho, p = stats.spearmanr(df[col], df['consensus_dice'])
    print(f"{label} Rho: {rho:.3f} (p={p:.3f})")
```
